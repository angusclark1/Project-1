```python
import os
import requests
import numpy as pd 
import pandas as pd
import matplotlib.pyplot as plt
from dotenv import load_dotenv
import yfinance as yf 
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation

%matplotlib inline
```

```python
load_dotenv()
```

```python
AAPL = yf.Ticker('AAPL')
SQ = yf.Ticker('SQ')
NU = yf.Ticker('NU')
LIT = yf.Ticker('LIT')
URNM = yf.Ticker('URNM')
```

```python
aapl_hist = AAPL.history(period='5y')
sq_hist = SQ.history(period='5y')
nu_hist = NU.history(period='5y')
lit_hist = LIT.history(period='5y')
urnm_hist = URNM.history(period='5y')
```

```python
print(aapl_hist)
print(sq_hist)
print(nu_hist)
print(lit_hist)
print(urnm_hist)
```

```python
df_aapl = pd.DataFrame(aapl_hist)
df_sq = pd.DataFrame(sq_hist)
df_nu = pd.DataFrame(nu_hist)
df_lit = pd.DataFrame(lit_hist)
df_urnm = pd.DataFrame(urnm_hist)
```

```python
print(df_aapl)
print(df_sq)
print(df_nu)
print(df_lit)
print(df_urnm)
```

```python
df_aapl['Ticker']='AAPL'
df_sq['Ticker']='SQ'
df_nu['Ticker']='NU'
df_lit['Ticker']='LIT'
df_urnm['Ticker']='URNM'
```

```python
df_aapl = df_aapl.drop(columns=["High","Low","Volume","Dividends","Stock Splits"])
df_sq = df_sq.drop(columns=["High","Low","Volume","Dividends","Stock Splits"])
df_nu = df_nu.drop(columns=["High","Low","Volume","Dividends","Stock Splits"])
df_lit = df_lit.drop(columns=["High","Low","Volume","Dividends","Stock Splits"])
df_urnm = df_urnm.drop(columns=["High","Low","Volume","Dividends","Stock Splits"])
```

```python
df_aapl.to_csv('data_aapl.csv')
df_sq.to_csv('data_sq.csv')
df_nu.to_csv('data_nu.csv')
df_lit.to_csv('data_lit.csv')
df_urnm.to_csv('data_urnm.csv')
```
