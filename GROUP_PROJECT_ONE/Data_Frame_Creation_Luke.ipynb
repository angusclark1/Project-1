```python
data_aapl_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_aapl.csv')
data_brk_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_brk.csv')
data_btc_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_btc.csv')
data_cba_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_cba.csv')
data_dis_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_dis.csv')
data_eth_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_eth.csv')
data_gld_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_gld.csv')
data_jnj_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_jnj.csv')
data_ko_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_ko.csv')
data_lit_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_lit.csv')
data_ma_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_ma.csv')
data_mc_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_mc.csv')
data_nu_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_nu.csv')
data_pg_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_pg.csv')
data_rivn_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_rivn.csv')
data_sq_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_sq.csv')
data_tsla_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_tsla.csv')
data_urnm_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_urnm.csv')
data_usdt_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_usdt.csv')
data_voo_path = Path('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_voo.csv')
```

```python
data_aapl = pd.read_csv(data_aapl_path)
data_brk = pd.read_csv(data_brk_path)
data_btc = pd.read_csv(data_btc_path)
data_cba = pd.read_csv(data_cba_path)
data_dis = pd.read_csv(data_dis_path)
data_eth = pd.read_csv(data_eth_path)
data_gld = pd.read_csv(data_gld_path)
data_jnj = pd.read_csv(data_jnj_path)
data_ko = pd.read_csv(data_ko_path)
data_lit = pd.read_csv(data_lit_path)
data_ma = pd.read_csv(data_ma_path)
data_mc = pd.read_csv(data_mc_path)
data_nu = pd.read_csv(data_nu_path)
data_pg = pd.read_csv(data_pg_path)
data_rivn = pd.read_csv(data_rivn_path)
data_sq = pd.read_csv(data_sq_path)
data_tsla = pd.read_csv(data_tsla_path)
data_urnm = pd.read_csv(data_urnm_path)
data_usdt = pd.read_csv(data_usdt_path)
data_voo = pd.read_csv(data_voo_path)
```

```python
group_data = pd.concat([data_aapl,data_brk,data_btc,data_cba,data_dis,data_eth,data_gld,data_jnj,data_ko,data_lit,data_ma,data_mc,data_nu,data_pg,data_rivn,data_sq,data_tsla,data_urnm,data_usdt,data_voo], axis="rows", join="inner")
group_data.head()
```

```python
group_data = group_data.set_index(pd.to_datetime(group_data["Date"], infer_datetime_format=True))
group_data = group_data.drop(columns=["Date"])
```
