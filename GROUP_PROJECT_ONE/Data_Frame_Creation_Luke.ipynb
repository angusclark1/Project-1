```python
import yfinance as yf
import pandas as pd
```

```python
#CODE FOR TICKER:DIS
```

```python
dis = yf.Ticker("dis")
```

```python
hist_dis = dis.history(period="5y")
```

```python
print(hist_dis)
```

```python
df_dis = pd.DataFrame(hist_dis) 
df_dis.head()
```

```python
df_dis["Ticker"]="DIS"
df_dis.head()
```

```python
df_dis = df_dis.drop(columns=["High","Low","Volume","Dividends","Stock Splits"])
df_dis.head()
```

```python
first_column_dis = df_dis.pop('Ticker')
  
df_dis.insert(0, 'Ticker', first_column_dis)
  
df_dis.head()
```

```python
df_dis.to_csv('/Users/luketarlinton/Desktop/Jupiter-Workspace/GROUP_PROJECT_ONE/CSVs/data_dis.csv')
```
